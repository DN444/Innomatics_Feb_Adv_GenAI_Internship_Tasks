```python
# 1
class SmartLight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"
    def control(self,action):
        if action.upper()=="ON":
            self.status="ON"
        else:
            self.status="OFF"
    def display(self):
        print(f"{self.name} currently {self.status}")
name=input()
action=input()
light=SmartLight(name)
light.control(action)
light.display()
```

```python
# 2
class Employee:
    def __init__(self,name,emp_id,department):
        self.name=name
        self.emp_id=emp_id
        self.department=department
    def id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
name=input()
emp_id=input()
department=input()
employee=Employee(name,emp_id,department)
employee.id_card()
```

```python
# 3
class Contact:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name=input()
phone=input()
contact=Contact(name,phone)
contact.display()
```

```python
# 4
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display_tag(self):
        print("Product:", self.name)
        print("Price: Rupees " + str(self.price))
name=input()
price=input()
product=Product(name,price)
product.display_tag()
```

```python
# 5
class Movie:
    def __init__(self,name,director,rating):
        self.name=name
        self.director=director
        self.rating=rating
    def display(self):
        print("Movie:",self.name)
        print("Directed by:",self.director)
        print("Rating:",self.rating,"/ 10")
name=input()
director=input()
rating=input()
movie=Movie(name,director,rating)
movie.display()
```

```python
# 6
class Delivery:
    def __init__(self,customer,address,item,quantity):
        self.customer=customer
        self.address=address
        self.item=item
        self.quantity=quantity
    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)
        print("Item:", self.item)
        print("Quantity:", self.quantity)
customer=input()
address=input()
item=input()
quantity=input()
delivery=Delivery(customer,address,item,quantity)
delivery.display()
```
