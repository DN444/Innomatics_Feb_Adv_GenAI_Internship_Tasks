```python
# Problem 1
sent=input("Enter sentence:")
words=sent.split()
unique=set(words)
print("Unique words count:", len(unique))
print("Unique words:", unique)
```

```python
# Problem 2
employees={}
n=int(input("Enter number of employees:"))
for i in range(n):
    name=input("Enter employee name:")
    sal=int(input("Enter salary:"))
    employees[name]=sal
highemp=None
highsal=0
for name,salary in employees.items():
    if salary>highsal:
        highsal=salary
        highemp=name
print(f"Highest Salary: {highemp} - {highsal}")
```

```python
# Problem 3
numbers=list(map(int,input("Enter numbers:").split()))
maxval=numbers[0]
minval=numbers[0]
for i in numbers:
    if i>maxval:
        maxval=i
    if i<minval:
        minval=i
print("List:",numbers)
print("Max:",maxval)
print("Min:",minval)
```

```python
# Problem 4
prices=list(map(int,input("Enter product prices:").split()))
count=0
for i in prices:
    if i>1000:
        count+=1
print("Products above 1000:", count)
```

```python
# Problem 5
attendance = input("Enter attendance (P/A): ").split()
present_count=0
for status in attendance:
    if status.upper()=="P":
        present_count+=1
percentage=(present_count/len(attendance))*100
print("Attendance Percentage:", percentage)
```

```python
# Problem 6
phone_numbers=list(map(int,input("Enter phone numbers:").split()))
unique_numbers=set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7
text=input("Enter a string: ")
freq={}
for char in text:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print(freq)
```

```python
# Problem 9
employees={}
n=int(input("Enter number of employees:"))
for i in range(n):
    name=input("Enter employee name:")
    salary=int(input("Enter salary:"))
    employees[name]=salary
key=input("Enter employee name to check:")
if key in employees:
    print("Employee exists")
else:
    print("Employee doesn't exist")
```

```python
# Problem 10
marks=list(map(int,input("Enter marks separated by space:").split()))
tot=0
for mark in marks:
    tot+=mark
average=tot/len(marks)
print("Average Marks:", average)
```
