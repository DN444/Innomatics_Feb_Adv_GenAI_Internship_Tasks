```python
#Program 1
userinput=input("Enter username:")
passinput=input("Enter password:")
username="admin"
password="1234"
if userinput==username and passinput==password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Program 2
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for mark in marks:
    if mark>=50:
        pass_count+=1
    else:
        fail_count+=1
print("Total Students:",len(marks))
print("Passed Students:",pass_count)
print("Failed Students:",fail_count)
```

```python
#Program 3
names=[" Alice ","bob"," CHARLIE "]
cleaned=[]
for name in names:
    cname=name.strip().lower()
    cleaned.append(cname)
print("Cleaned Names:",cleaned)
```

```python
#Program 4 
messages=["Hi","Welcome to the platform","OK"]
for message in messages:
    length=len(message)
    print(f"Message: '{message}' has length: {length}")
    if length>10:
        print("This message is longer than 10 characters.")
```

```python
#Program 5
logs=["INFO","ERROR","WARNING","ERROR"]
errcount=0
for log in logs:
    if log=="ERROR":
        errcount += 1
print("Total ERROR entries:",errcount)
```
