```python
# 1
def manage_parking(capacity,logs):
    parked=0
    peak=0
    for log in logs:
        if log.lower()=="in":
            parked+=1
        elif log.lower()=="out" and parked>0:
            parked-=1
        if parked>peak:
            peak=parked
    if parked>capacity:
        status="Full"
    else:
        status="Available"
    return parked,peak,status
capacity=int(input())
logs=input().split()
current,peak,status=manage_parking(capacity, logs)
print("Currently Parked Vehicles:", current)
print("Parking Status:", status)
```

```python
# 2
def estimate_time(distance,traffic,weather):
    time=distance*5
    if traffic.lower()=="high":
        time+=15
    elif traffic.lower()=="medium":
        time+=10
    if weather.lower()=="rainy":
        time+=10
    elif weather.lower()=="stormy":
        time+=20
    return time
distance=float(input())
traffic=input()
weather=input()
eta=estimate_time(distance, traffic, weather)
print("Estimated Delivery Time:", int(eta), "minutes")
```

```python
# 3
def analyze_occupancy(total_seats,bookings):
    booked=0
    for seat in bookings:
        booked+=seat
    occupancy=(booked/total_seats)*100
    if occupancy==100:
        status="Housefull"
    elif occupancy>=75:
        status="Almost Full"
    else:
        status="Available"
    return occupancy,status
total_seats=int(input())
bookings=list(map(int,input().split()))
occupancy,status=analyze_occupancy(total_seats,bookings)
print("Occupancy:", int(occupancy), "%")
print("Show Status:", status)
```

```python
# 4
def classify_load(readings):
    total=0
    for value in readings:
        total+=value
    average=total/len(readings)
    if average<50:
        status="Normal"
    elif average<=80:
        status="Warning"
    else:
        status="Critical"
    return average,status
readings=list(map(int,input().split()))
average, status=classify_load(readings)
print("Average CPU Load:", int(average), "%")
print("Server Status:", status)
```

```python
# 5
def monitor_resources(resources):
    overused=[]
    for resource,hours in resources.items():
        if hours>8:
            overused.append(resource)
    if len(overused)>0:
        alert="Yes"
    else:
        alert="No"
    return overused,alert
n=int(input())
resources={}
for i in range(n):
    name=input()
    hours=int(input())
    resources[name]=hours
overused,alert=monitor_resources(resources)
print("Overused Resources:", ", ".join(overused) if overused else "None")
print("Energy Alert:", alert)
```

```python
# 6
def manage_registration(capacity,registrations):
    confirmed=0
    for i in range(registrations):
        if confirmed<capacity:
            confirmed+=1
    waitlisted=registrations-confirmed
    if registrations>=capacity:
        status="Closed"
    else:
        status="Open"
    return confirmed,waitlisted,status
capacity=int(input())
registrations=int(input())
confirmed,waitlisted,status=manage_registration(capacity,registrations)
print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
