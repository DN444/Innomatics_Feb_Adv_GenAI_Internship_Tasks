```python
# 1
def validate_recharge(amount,plan):
    valid_plans=[199, 299, 399, 599]
    if amount>=50 and plan in valid_plans:
        return True
    return False
while True:
    amount=int(input())
    plan=int(input())
    if validate_recharge(amount,plan):
        print("Recharge Successful")
        break
    else:
        print("Invalid Recharge. Try Again")
```

```python
# 2
def check_inventory(products):
    for product,stock in products.items():
        if stock<15:
            print(product,"- Reorder Alert")
        else:
            print(product,"- Stock OK")
n=int(input())
products={}
for i in range(n):
    name=input()
    stock=int(input())
    products[name]=stock
check_inventory(products)
```

```python
# 3
def process_result(marks):
    total=0
    for mark in marks:
        total+=mark
    average=total/len(marks)
    if average>=50:
        return "Pass"
    else:
        return "Fail"
marks=list(map(int,input().split()))
result=process_result(marks)
print(result)
```

```python
# 4
def calculate_fare(distance,peak):
    fare=50+(12*distance)
    if peak.lower()=="yes":
        fare+=fare*0.25
    return fare
while True:
    distance=float(input())
    peak=input()
    fare=calculate_fare(distance,peak)
    print("Total Fare:", fare)
    retry=input()
    if retry.lower()!="yes":
        break
```

```python
# 5
def check_eligibility(attendance):
    present=0
    for status in attendance:
        if status.upper()=="P":
            present+=1
    percentage=(present/len(attendance))*100
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance=input().split()
result=check_eligibility(attendance)
print(result)
```

```python
# 6
def check_password(password):
    if len(password)<8:
        return "Weak Password"
    has_digit=False
    has_special=False
    for char in password:
        if char.isdigit():
            has_digit=True
        if char in "@#$":
            has_special=True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
password = input()
print(check_password(password))
```
