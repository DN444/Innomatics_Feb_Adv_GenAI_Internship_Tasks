```python
# 1
def analyze_engagement(likes):
    total=0
    for like in likes:
        total+=like
    if total>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"
    return total,status
likes=list(map(int,input().split()))
total,status=analyze_engagement(likes)
print("Total Likes:", total)
print("Post Status:", status)
```

```python
# 2
def check_stock(stock):
    if stock<10:
        return "Low Stock Alert"
    else:
        return "Stock Sufficient"
stock=int(input())
status=check_stock(stock)
print("Medicine Stock:", stock)
print("Status:", status)
```

```python
# 3
def rainfall_status(data,required):
    total=0
    for value in data:
        total+=value
    average=total/len(data)
    if average>=required:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    return average,status
data=list(map(int,input().split()))
required=float(input())
average,status=rainfall_status(data,required)
print("Average Rainfall:", average)
print("Rainfall Status:", status)
```

```python
# 4
def check_duplicates(usernames):
    if len(usernames)!=len(set(usernames)):
        return "Yes"
    else:
        return "No"
usernames=input().split()
result=check_duplicates(usernames)
print("Duplicate Accounts Found:", result)
```

```python
# 5
def check_eligibility(age):
    if age>=18:
        return "Eligible"
    else:
        return "Not Eligible"
age=int(input())
status=check_eligibility(age)
print("Patient Age:", age)
print("Eligibility Status:", status)
```

```python
# 6
def filter_premium(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    return premium
prices=list(map(int,input().split()))
premium=filter_premium(prices)
print("Premium Crops:", premium)
```

```python
# 7
def system_health(errors):
    if errors==0:
        return "Healthy"
    elif errors<=5:
        return "Minor Issues"
    else:
        return "Critical Issues"
errors=int(input())
status=system_health(errors)
print("Error Count:", errors)
print("System Status:", status)
```

```python
# 8
def transaction_status(amount):
    if amount<=50000:
        return "Approved"
    else:
        return "Rejected"
amount=float(input())
status=transaction_status(amount)
print("Transaction Amount:", int(amount))
print("Transaction Status:", status)
```

```python
# 9
def attendance_check(attendance):
    present=0
    for day in attendance:
        if day.upper()=="P":
            present+=1
    percentage=(present/len(attendance))*100
    if percentage>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    return percentage,status
attendance=input().split()
percentage,status=attendance_check(attendance)
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", status)
```

```python
# 10
def calculate_bill(units):
    if units<=100:
        bill=units * 3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)
    if bill<500:
        usage="Low Usage"
    elif bill<=1500:
        usage="Moderate Usage"
    else:
        usage="High Usage"
    return bill,usage
units=int(input())
bill,usage=calculate_bill(units)
print("Total Bill:", bill)
print("Usage Status:", usage)
```
